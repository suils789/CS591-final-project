```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv',encoding = 'latin1')
```

```python
data
```

```python
aid = data['app_id'].values.tolist()
rating = data['star_rating'].values.tolist()
rid = data['reviewer'].values.tolist()
```

```python
items = data.groupby('app_id').groups
users = data.groupby('reviewer').groups
```

```python
import networkx as nx
from networkx.algorithms import bipartite
B=nx.Graph()
B.add_nodes_from(list(items.keys()), bipartite=0)
B.add_nodes_from(list(users.keys()), bipartite=1)
for i in range(len(rid)):
    try:
        B.add_edge(aid[i], rid[i], weight=rating[i])
    except:
        pass
```

```python
from networkx.algorithms.bipartite import biadjacency_matrix

r_order = list(set(aid))
c_order = list(set(rid))
mat = biadjacency_matrix(B, row_order = r_order, column_order = c_order)
```

```python
mat
```

```python
from collections import defaultdict
globalAverage = data['star_rating'].mean()
```

```python
userRatings = defaultdict(list)
itemRatings = defaultdict(list)
for i in range(len(rid)):
    user = rid[i]
    item = aid[i]
    userRatings[user].append(rating[i])
    itemRatings[item].append(rating[i])
userAverage = {}
for u in userRatings:
    userAverage[u] = sum(userRatings[u]) / len(userRatings[u])
itemAverage = {}
for i in itemRatings:
    itemAverage[i] = sum(itemRatings[i]) / len(itemRatings[i])
```

```python
def find_non_zero_entry(m):
    l1 = m.nonzero()[0]
    l2 = m.nonzero()[1]
    l = []
    for i in range(len(l1)):
        l.append((l1[i],l2[i]))
    return l
l = find_non_zero_entry(mat)
```

```python
Y = []
userAvg = []
itemAvg = []
userLoyalty = []
itemPopularity = []
for pair in l:
    Y.append(mat[pair[0], pair[1]])
    userAvg.append(sum(userRatings[c_order[pair[1]]])/len(userRatings[c_order[pair[1]]]))
    itemAvg.append(sum(itemRatings[r_order[pair[0]]])/len(itemRatings[r_order[pair[0]]]))
    userLoyalty.append(len(userRatings[c_order[pair[1]]]))
    itemPopularity.append(len(itemRatings[r_order[pair[0]]]))
```

```python
df_train = pd.DataFrame(np.array([Y,userAvg,itemAvg, userLoyalty, itemPopularity]).T, columns=['Y', 'userAvg', 'itemAvg', 'userLoyalty', 'itemPopularity'])
import statsmodels.api as sm
model = sm.OLS(df_train['Y'], df_train[['userAvg', 'itemAvg', 'userLoyalty', 'itemPopularity']]).fit()

print(model.summary())
```

```python
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train[['X1', 'X2', 'X3', 'X4']], df_train['Y'], test_size=0.3, random_state=1)
lr = LogisticRegression(C=1000.0, random_state=0)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
lr_avg_error = ((y_test - y_pred)*(y_test - y_pred)).sum()/len(y_pred)
print('Logistic regression average error: %.6f'% lr_avg_error)
print('Number of misclassified samples: %d' % (y_test != y_pred).sum())
from sklearn.metrics import accuracy_score

print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))    

print('Coefficients: \n', lr.coef_)
```

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
y_train_pred = regr.predict(X_train)
y_test_pred = regr.predict(X_test)

regr_avg_error = (((y_test - y_test_pred)*(y_test - y_test_pred)).sum())/len(y_test_pred)
print('Linear regression average error: %.6f'% regr_avg_error)
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print('MSE train: %.3f, test: %.3f' %(mean_squared_error(y_train, y_train_pred),
                                      mean_squared_error(y_test, y_test_pred)))
 
print('R^2 train: %.3f, test: %.3f' %(r2_score(y_train, y_train_pred),
                                      r2_score(y_test, y_test_pred)))
print('Coefficients: \n', regr.coef_)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(criterion='entropy',
                                n_estimators=100, 
                                random_state=1,
                                n_jobs=2)
forest.fit(X_train, y_train)

y_pred_forest = forest.predict(X_test)
rf_avg_error = ((y_test - y_pred_forest)*(y_test - y_pred_forest)).sum()/len(y_pred_forest)
print('RF Misclassified samples: %d'% (y_test != y_pred_forest).sum())
print('RF average error: %.6f'% rf_avg_error)
from sklearn.metrics import accuracy_score

print('RF Accuracy: %.2f'% accuracy_score(y_test, y_pred_forest))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred1 = gnb.predict(X_test)
nb_avg_error = ((y_test - y_pred1)*(y_test - y_pred1)).sum()/len(y_pred1)
print('Naive Bayes Misclassified samples: %d'% (y_test != y_pred1).sum())
print('Naive Bayes average error: %.6f'% nb_avg_error)
from sklearn.metrics import accuracy_score
print('Naive Bayes Accuracy: %.2f'% accuracy_score(y_test, y_pred1))
```

```python
def matrix_factorization(R, P, Q, K, entries, steps=2000, alpha=0.0002, beta=0.02):
    Q = Q.T
    for step in range(steps):
        for entry in entries:
            i = entry[0]
            j = entry[1]
            eij = R[i, j] - np.dot(P[i,:],Q[:,j])
            for k in range(K):
                P[i][k] = P[i][k] + alpha * (2 * eij * Q[k][j] - beta * P[i][k])
                Q[k][j] = Q[k][j] + alpha * (2 * eij * P[i][k] - beta * Q[k][j])
        e = 0
        for entry in entries:
            i = entry[0]
            j = entry[1]
            e = e + pow(R[i,j] - np.dot(P[i,:],Q[:,j]), 2)
            for k in range(K):
                e = e + (beta/2) * ( pow(P[i][k],2) + pow(Q[k][j],2) )
        if e < 0.001:
            break
    return P, Q.T
```
